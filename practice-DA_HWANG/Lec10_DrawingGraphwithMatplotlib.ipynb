```python
import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [10, 20, 30, 40]) # x축값, y축값
plt.title('plotting') # title 설정
plt.show()
```

```python
import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [10, 20, 30, 40, 50]) #x축과 y축의 데이터 개수가 같지 않으면 오류가 발생함.
plt.show()
```

```python
import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [10, 20 ,30, 40])
plt.title('plotting')
plt.xlabel('x axis') # x축의 이름 지정
plt.ylabel('y axis') # y축의 이름 지정
plt.show()
```

```python
import matplotlib.pyplot as plt 

plt.plot([1, 2, 3, 4], [10, 20 ,30, 40], label='ascending') #label 속성을 통해 레이블을 출력함
plt.plot([1, 2, 3, 4], [40, 30, 20 ,10], label='descending')
plt.title('legend')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend() # 범례를 출력
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # 'ro' = red를 뜻함
plt.axis((0, 6, 0, 20)) # plt.axis=[xmin, xmax, ymin, ymax] 로 각각 x축의 최소, 최대, y축의 최소, 최대를 지정
plt.show()
```

색상

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color = 'skyblue', label='skyblue')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'pink', label ='pink') # color 속성명(키워드 인자) 표시는 필수는 아님.
plt.title('color')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

선 스타일

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color = 'r', linestyle='--', label='dashed')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'b', ls=':', label='dotted')
plt.title('linestyle')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

마커

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color = 'r', linestyle='--', marker = 'd', label='dashed')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'b', ls=':', marker='^', label='dotted')
plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

선 굵기

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40], color = 'r', linestyle='--', marker = 'd', linewidth=3, label='dashed')
plt.plot([1, 2, 3, 4], [40, 30, 20, 10], 'b', ls=':', marker='^', linewidth =0.5, label='dotted')
plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
# 라인스타일이 없으면 마커만 찍힘
plt.plot([10, 20, 30, 40], 'r.', label='circle')
plt.plot([40, 30, 20, 10], 'g^', label = 'triangle up')

plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 30 ,40], '-.m*', [1, 2, 3, 4,], [50, 60, 70, 80], 'r.', label ='circle')
plt.plot([40, 30, 20, 10], 'g^', label='triangle up')

plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c ="b",
         lw = 5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("Style")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], "-.bo", lw=3, ms=8, mec="skyblue", mew=2, mfc="y")
plt.title("Styel")
plt.show()
```
