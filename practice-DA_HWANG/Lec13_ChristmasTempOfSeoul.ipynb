연도별 평균 기온 

```python
import csv
import matplotlib.pyplot as plt

f = open("seoul.csv", encoding='cp949') # seoul.csv파일을 불러오고
data = csv.reader(f) #reader메소드를 통해 파일을 읽는다.

header = next(data) # 헤더를 출력

temp_day =[] # 배열을 만든 후 
temp_mean =[]
year =[]

for row in data:

    if int(row[0].split('-')[0]) > 1907: # 날짜를 분리한 후 연도가 1907년 이상인 데이터 중 
        if row[2] !='': # 평균기온이 공값이 아니면
            temp_day.append(float(row[2])) # temp_day 배열에 추가
            if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '31'):
            # 12월 31일에 해당하는 데이터라면 
                year.append(int(row[0].split('-')[0])) # 연도 배열에 추가하고
                temp_mean.append(sum(temp_day) / len(temp_day)) # 평균기온 데이터의 평균값을 계산
                temp_day =[] # 다음년도의 평균 기온 데이터를 1월 1일부터 새롭게 저장해야하므로 
                #초기화 해준다.    
f.close()

plt.figure(figsize=(10, 4))
plt.rc('font', family='Malgun Gothic')
plt.title('평균 기온 변화 그래프')
plt.plot(year, temp_mean, color='k', marker='d')
plt.xlabel('year')
plt.ylabel('Average Temperature')
plt.grid()
plt.show()
```

추세선 = 예측선

```python
import matplotlib.pyplot as plt
import numpy as np

#키와 몸무게 배열 데이터(예시)
height = [160, 162, 165, 168, 170, 172, 175, 178, 180, 182]
weight = [55, 57, 70, 72, 65, 68, 70, 73, 83, 88]

degree = 1 # 1차 다항식을 의미
coefficients = np.polyfit(height, weight, degree) # polyfit(x, y, n차함수)
#coefficients는 배열로 저장되며, [0]에는 기울기, [1]에는 절편이 저장된다.
polynomial = np.poly1d(coefficients) # poly1d(coe) => coe[0]x + coe[1]
#polynomial에는 함수가 저장된다.
height_range = np.linspace(min(height), max(height), 100) # 추세선을 위한 x값 생성

#한글 처리
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

#그래프 그리기
plt.scatter(height, weight, label='데이터 포인트') # scatter(x, y, label이름)
plt.plot(height_range, polynomial(height_range), color='red', label='추세선') # plot(x, y, color, label이름 etc)

plt.xlabel('키 (cm)') # x레이블
plt.ylabel('몸무게 (kg)') # y레이블 
plt.title('키와 몸무게의 추세선') #제목
plt.legend() # 범례 추가

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

temp_day =[]
temp_mean =[]
year =[]

for row in data:
    if int (row[0].split('-')[0]) > 1907:
        if row[2] !='':
            temp_day.append(float(row[2]))
            if row[0].split('-')[1]=='12' and row[0].split('-')[2] =='31' :
                year.append(int(row[0].split('-')[0]))
                temp_mean.append(sum(temp_day) / len(temp_day))
                temp_day = []

f.close()

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False


plt.figure(figsize = (10, 4))
plt.scatter(year, temp_mean, color='b', marker='*')

coeff = np.polyfit(year, temp_mean, 1)
poly = np.poly1d(coeff)
plt.plot(year, poly(year), '--r')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.grid() # 그리드(격자 출력)
plt.title('평균 기온 변화 그래프의 추세선')
plt.show()
print(f"추세선 기울기 : {coeff[0]}")
```
