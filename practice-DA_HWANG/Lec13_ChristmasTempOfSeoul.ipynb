```python
import csv 
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if(row[2] !='') and (row[3] !='') and (row[4] !=''):
        if(row[0].split('-')[1] =='12') and (row[0].split('-')[2] =='25'):
            mean.append(float(row[2])) #
            low.append(float(row[3])) #
            high.append(float(row[4])) #
f.close()

plt.figure(figsize=(10, 2))
plt.plot(mean, 'k')
plt.plot(low, 'b')
plt.plot(high, 'r')
plt.xlabel('Day (December 25)')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open("seoul.csv", encoding='cp949') # seoul.csv 
data = csv.reader(f) #reader   .

header = next(data) #  

temp_day =[] #    
temp_mean =[]
year =[]

for row in data:

    if int(row[0].split('-')[0]) > 1907: #     1907    
        if row[2] !='': #   
            temp_day.append(float(row[2])) # temp_day  
            if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '31'):
            # 12 31   
                year.append(int(row[0].split('-')[0])) #   
                temp_mean.append(sum(temp_day) / len(temp_day)) #    
                temp_day =[] #     1 1   
                # .    
f.close()

plt.figure(figsize=(10, 4))
plt.rc('font', family='Malgun Gothic')
plt.title('   ')
plt.plot(year, temp_mean, color='k', marker='d')
plt.xlabel('year')
plt.ylabel('Average Temperature')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#   ()
height = [160, 162, 165, 168, 170, 172, 175, 178, 180, 182]
weight = [55, 57, 70, 72, 65, 68, 70, 73, 83, 88]

degree = 1 # 1  
coefficients = np.polyfit(height, weight, degree) # polyfit(x, y, n)
#coefficients  , [0] , [1]  .
polynomial = np.poly1d(coefficients) # poly1d(coe) => coe[0]x + coe[1]
#polynomial  .
height_range = np.linspace(min(height), max(height), 100) #   x 

# 
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

# 
plt.scatter(height, weight, label=' ') # scatter(x, y, label)
plt.plot(height_range, polynomial(height_range), color='red', label='') # plot(x, y, color, label etc)

plt.xlabel(' (cm)') # x
plt.ylabel(' (kg)') # y 
plt.title('  ') #
plt.legend() #  

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

temp_day =[]
temp_mean =[]
year =[]

for row in data:
    if int (row[0].split('-')[0]) > 1907:
        if row[2] !='':
            temp_day.append(float(row[2]))
            if row[0].split('-')[1]=='12' and row[0].split('-')[2] =='31' :
                year.append(int(row[0].split('-')[0]))
                temp_mean.append(sum(temp_day) / len(temp_day))
                temp_day = []

f.close()

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False


plt.figure(figsize = (10, 4))
plt.scatter(year, temp_mean, color='b', marker='*')

coeff = np.polyfit(year, temp_mean, 1)
poly = np.poly1d(coeff)
plt.plot(year, poly(year), '--r')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.grid() # ( )
plt.title('    ')
plt.show()
print(f"  : {coeff[0]}")
```
