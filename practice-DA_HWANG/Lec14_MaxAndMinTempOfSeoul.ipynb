```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
result = []

for row in data:
    if row[4] != '':
        result.append(float(row[4])) # 최고기온

plt.hist(result, bins=100, color='r') # 최고기온을, 99등분하여 빨간색 막대그래프로 나타냄.
plt.show()
```

```python
a = []
for i in range(5): # 반복
    a.append(i)
print(a)

b = [i for i  in range(5)] # 컴프리헨션
print(b)
```

```python
print([i**2 for i  in range(5)])
print([pow(i, 3) for i in range(5)]) #pow(i, n) i룰 n제곱함
```

```python
print([[i+1] for i in range(5)]) # 1씩 증가하는 [i]를 컴프리헨션함.
print([(i+1, ) for i  in range(5)]) # 1씩 증가하는 (i+1,)를 컴프리헨션함.
print([{i: i**2} for i in range(5)]) # 1씩 증가하는 {i : i**2}를 컴프리헨션함.
print([{i+1} for i in range(5)]) # 1씩 증가하는 {i}를 컴프리헨션함.
```

```python
print([i for i  in range(6)])# 1씩 증가하는 i를 컴프리헨션
print([i for i  in range(6) if i%2 ==0])# i가 짝수일 경우에만 배열에 저장
print([i for i in range(6) if i %2 ==1])# i가 홀수일 경우에만 배열에 저장
```

```python
age = 10
adult ='성인' if age >= 18 else '미성년' # value1 = if intvalue >= n else value2 형태가 가능
print(adult)
```

```python
print([0 if i%2 ==0 else 1 for i  in range(10)])
#i를 1씩 10까지 증가시키며, i가 짝수면 0, i가 홀수면 1을 저장
print(['even' if i%2 ==0 else'odd' for i in range(10)])
#위와 같은 조건이나, 0을 even, 1을 odd로 변경
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
aug =[]

for row in data:
    month = row[0].split('-')[1]
    if row[4] !='': #최고기온이 존재할경우
        if month =='08': #월이 8월일 경우
            aug.append(float(row[4]))

plt.hist(aug, bins=100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
aug =[]
jan =[]

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))

plt.figure(dpi=150)
plt.hist(aug, bins =100, color ='r', label ='Aug')
plt.hist(jan, bins =100, color ='b', label ='Jan')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

month = [[] for i  in range(12)] #월별 데이터를 저장2

for row in data :
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        # 인덱싱은 0부터 시작되므로 해당월에서 1을 빼준다.
plt.figure(figsize=(8,5), dpi=200)
plt.boxplot(month)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

day = [[] for i  in range(31)]
for row in data:
    if row[-1] !='':
        if row[0].split('-')[1] == '01' : # 1월
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            # 날짜별로 
plt.figure(figsize=(10, 5), dpi=300)
plt.boxplot(day, showfliers=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [3, 4, 5, 6 ,7]
plt.boxplot(data, showmeans=True)
print(f'중앙값(medial): {np.median(data)}')
print(f'평균값(mean): {np.median(data)}')
plt.show()
```
