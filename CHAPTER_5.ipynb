### 그래프의 여러 속성

```python
import matplotlib.pyplot as plt

# 데이터 생성
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 선 플롯 생성
plt.plot(x, y, marker='o', linestyle='-', color='b', label='소수(Prime Numbers)')
# marker는 데이터 포인트에 사용될 마커 모양을 지정한다. 'o'는 원형 마커를 의미하며, '^'(삼각형), 's'(사각형), 'x'(엑스) 등이 있다.
# linestyle은 데이터 포인트 간의 선 스타일을 지정한다. '-'는 실선을 의미하며, ':'(점선), '--'(대시선), '-.'(점 대시) 등이 있다.
# color는 그래프의 선과 마커 색상을 지정한다. 'b'는 파란색을 의미하며, 다른 색상의 기본옵션은 'k'(검정색)을 제외하고는 각 색상의 첫 글자와 같으며, 16진수 색상코드나 RGB값으로도 색을 지정할 수 있다.
# label은 범례에 표시될 텍스트로, 해당 선이 무엇을 나타내는지 설명한다.

# 축 및 제목 설정
plt.xlabel('X-축(axis)')
plt.ylabel('Y-축(axis)')
plt.title('소수 그리기(Prime Numbers Plot)')
plt.legend() # 범례 표시 color는 그래프의 선과 마커 색상을 지정한다. 'b'는 파란색을 의미하며, 다른 색상의 기본옵션은 'k'(검정색)을 제외하고는 각 색상의 첫 글자와 같으며, 16진수 색상코드나 RGB값으로도 색을 지정할 수 있다.
# label은 범례에 표시될 텍스트로, 해당 선이 무엇을 나타내는지 설명한다.

# 축 및 제목 설정
plt.xlabel('X-축(axis)') # x축
plt.ylabel('Y-축(axis)') # y축
plt.title('소수 그리기(Prime Numbers Plot)') # 제목
plt.legend() # 범례 표시

# 플롯 표시
plt.show()
```

### 산점도

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(2025) # 같은 시드 값을 설정하면 난수 생성기를 통해 생성되는 난수들의 순서가 항상 동일하게 유지된다.

N = 80
x = np.random.rand(N) # 0과 1 사이의 난수 8개 생성
y = np.random.rand(N) # 0과 1 사이의 난수 8개 생성
colors = np.random.rand(N) # 0과 1 사이의 난수 8개 생성
area = (20* np.random.rand(N)) **2 # 0 에서 20 사이의 난수 N개 를 제곱한 결과를 area에 대입
plt.scatter(x, y, s=area, c=colors, alpha=0.7) # 인자 s는 크기  c 는 색상, alpha는 불투명도를 나타낸다.
plt.colorbar()
plt.show()
```

```python
x = np.random.randint(-100, 100, 1000) #1000개의 랜덤 정수 값 추출
y = np.random.randint(-100, 100, 1000) #1000개의 랜덤 정수 값 추출

mask1 = abs(x) > 50 # x에 저장된 값 중 50보다 큰 값만 추출
mask2 = abs(y) > 50 # y에 저장된 값 중 50보다 큰 값만 추출
x = x[mask1 + mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장 ,즉 +는 union을 뜻함
y = y[mask1 + mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장

plt.scatter(x, y, c=x, cmap='jet', alpha=0.4) # 산점도 그리기, cmap은 컬러맵을 뜻함
plt.colorbar() # 컬러바 표시
plt.show()
```

```python
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50
mask2 = abs(y) > 50

x = x[mask1 * mask2]
y = y[mask1 * mask2]

plt.scatter(x, y, c=x, cmap='jet', alpha=0.4)
plt.colorbar()
plt.show()
```

### 돗수분포표 그림 hist()

```python
plt.figure(figsize=(6, 4))
# plt.figure(figsize=(x, y))로 그림 크기를 지정함
plt.hist([1, 1, 1, 2, 2, 3, 4, 5, 6, 6, 6, 6], rwidth=.98, bins =[.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])
# plt.hist()로 돗수분포표를 그릴 수 있다. 첫번째 배열 인수의 값을 기준으로 배열 안의 값이 등장하는 빈도수로 그래프를 그림.
# bins는 히스토그램의 구간 경계를 지정하고, 각 구간에 해당하는 데이터 값의 개수가 히스토그램의 막대 높이로 나타난다.
# ex) [0.5, 1.5] 구간에는 값 1인 3개 포함된다.
plt.xticks(range(1, 7))
# x축을 1~6까지 출력한다.
plt.show()
```

```python
dice = np.random.choice(range(1, 7), 1000000, p = [0.15, 0.25, 0.3, 0.1, 0.1, 0.1])
#  1~6까지의 숫자가 써있는 주사위를 백만번 돌려 p는 1~6까지의 숫자가 나올 확률을 지정해준다.
plt.figure(figsize=(5, 3), dpi=200)
# 그림크키, dpi = 인치당 점.
plt.hist(dice, rwidth=.98, bins=np.arange(.5, 7))
#
plt.xticks(range(1, 7))

plt.savefig('주사위.png')
plt.show()

```

### 서브플롯 plt.subplot() 개요

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 4))

# 부분플롯 plt.subplot(m, n, index) m은 행, n은 열, index는 1부터 시작하는 부분플롯의 번호를 말하며, 왼쪽 상단이 우선이다.
plt.subplot(2, 3, 1)
plt.subplot(2, 3, 5)
plt.subplot(2, 3, 6)

plt.plot()
plt.show()
```

### 서브플롯 plt.subplots() 개요

```python
import matplotlib.pyplot as plt

plt.figure(figsize =(9, 4))

plt.subplots(2, 4) # plt.subplots()는 subplot과 다르게 figure를 꽉 채운 axes를 출력한다.
plt.show()
```

### 서브플롯 fig.add_subplot()개

```python
import matplotlib.pyplot as plt

# define figure
fig = plt.figure()

# add subplots
# (3,2,1)을 (321)로 줄여쓰는 것이 가능하다, 그러나 1~9까지만 가능
fig.add_subplot(321).set_title('321')
fig.add_subplot(322).set_title('322')
fig.add_subplot(323).set_title('323')
fig.add_subplot(324).set_title('324')
fig.add_subplot(325).set_title('325')
fig.add_subplot(326).set_title('326')

#adjust spaces
fig.subplots_adjust(wspace=.2, hspace=.7)
plt.show()
```

```python
fig = plt.figure(figsize=(7, 5))
#plt.figure()에 figsize 속성값까지 지정하여 변수에 대입도 가능하다.

ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
```
